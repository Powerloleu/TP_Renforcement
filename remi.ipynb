```python
import pandas as pd
import random
import numpy as np
```

```python
def dices(n_faces = 6, n_lancers = 5):
    return np.random.randint(low = 0, high = n_faces, size = n_lancers)
```

```python
simulations = pd.DataFrame([dices() for i in range(10000)])
simulations
```

```python
def prob_n_lancers(n_lancers = 5, num = 1, error = 0.05, precision = 0.01):
    N = (np.log(2) - np.log(error)) / (2 * precision**2)
    N = np.ceil(N).astype(int)
    simulations = pd.DataFrame([dices() for i in range(N)], columns = ['first', 'second', 'third', 'fourth', 'fifth'])
    return (simulations.isin([num]).any(axis = 1).sum())/simulations['first'].count()
```

# Algorithme d'iteration de valeurs

```python
def add_column(df, name, gamma):
    last_col = df.columns[-1]
    df[name] = df.iloc[:,0] + gamma*df.at[0, last_col]
    df[name].at[0] = df['V0'].at[0] + gamma* max([
        (1/2*df.at[1,last_col]+1/2*df.at[2,last_col]), 
        (1/10*df.at[3,last_col]+9/10*df.at[4,last_col])
        ])
    return df
```

```python
df = pd.DataFrame([0.0, 5.0, 100.0, 500.0, 3.0], columns = ['V0'])

for v in ['V1', 'V2', 'V3', 'V4', 'V5', 'V6']:
    df = add_column(df, v, 1)
df
```

```python
df = pd.DataFrame([0.0, 5.0, 100.0, 500.0, 3.0], columns = ['V0'])

for v in [f'V{i}' for i in range(1,12)]:
    df = add_column(df, v, 1/2)
df
```
